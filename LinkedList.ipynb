# Linked List

```python
##singly linked list-every node contain only one link of next node
##first node is head position and last node is tail position
##operation 1.add 2.delete 3.traversal
##Add/inserstion/delete at begin, end and inbetween.
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.ref = None
class Linkedlist:
    def __init__(self):
        self.head = None
    def print_LL(self):
        if self.head is None:
            print("linked list is empty")
        else:
            n=self.head
            while n is not None:
                print(n.data,"--->",end=' ')
                n=n.ref
    #addition of node          
    def add_begin(self,data):
        new_node=Node(data)
        new_node.ref=self.head
        self.head=new_node
        
    def add_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            n=self.head
            while n.ref is not None:
                n=n.ref
            n.ref=new_node  
    def add_after(self,data,x):
        n=self.head
        while n is not None:
            if x==n.data:
                break   
            n=n.ref
        if n is None:
            print("node is not present in LL")
        else:
            new_node=Node(data)
            new_node.ref=n.ref
            n.ref=new_node
    def add_before(self,data,x):
        if self.head.data==x:
            new_node=Node(data)
            new_node.ref=self.head
            self.head=new_node
        n=self.head
        while n is not None:
            if n.ref.data==x:
                break   
            n=n.ref
        if n.ref is None:
            print("Node is not found!")
        else:   
            new_node=Node(data)
            new_node.ref=n.ref
            n.ref=new_node 
    def insert_empty(self,data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("ll is not empty")
    #deletion of node        
    def delete_begin(self):
        if self.head is None:
            print('ll is empty')
        else:
            self.head=self.head.ref
    
    def delete_end(self):
        if self.head is None:
            print('ll is empty')
        elif self.head.ref is None:
            self.head=None
        else:
            n=self.head
            while n.ref.ref is not None:
                n=n.ref
            n.ref=None    
    def delete_by_value(self,x):
        if self.head is None:
            print('ll is empty')
            return
        else:
            if x==self.head.data:
                self.head=self.head.ref
                return
            n=self.head
            while n.ref is not None:
                if x==n.ref.data:
                        break
                n=n.ref      
            if n.ref is None:
                print('number not present')
            else:
                n.ref=n.ref.ref
            
LL1=Linkedlist()
LL1.add_begin(10)
LL1.add_end(100)
LL1.add_begin(20)
LL1.add_before(260,100)
LL1.add_after(240,10)
LL1.delete_begin()
LL1.delete_end()
LL1.delete_by_value(240)
LL1.print_LL()

            
```
